# Design Candidate Search

Run local perturbation search for a few baselines and inspect improvement distributions.

```python
from pathlib import Path

import pandas as pd

from copilot.search import search_candidates

df = pd.read_parquet('../data/processed/design_table.parquet')
df.head()
```

```python
baseline_ids = df['run_id'].head(3).tolist()
results = []
for run_id in baseline_ids:
    results.extend(search_candidates(df, int(run_id), num_candidates=100, top_n=3))
summary = pd.DataFrame([c.model_dump() for c in results])
summary[['candidate_id', 'baseline_run_id', 'predicted_cd', 'estimated_cd_improvement', 'feasibility_status', 'score']]
```

```python
ax = summary['estimated_cd_improvement'].hist(bins=12)
ax.set_title('Top candidate estimated Cd improvement')
ax.set_xlabel('baseline predicted Cd - candidate predicted Cd')
```
