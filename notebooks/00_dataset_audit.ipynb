# DrivAerML Dataset Audit

Quick audit of the processed design table before modeling.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

DATA = Path('../data/processed/design_table.parquet')
FIG_DIR = Path('../reports/figures/dataset_audit')
FIG_DIR.mkdir(parents=True, exist_ok=True)

df = pd.read_parquet(DATA)
df.head()
```

```python
print(df.shape)
pd.DataFrame({'column': df.columns, 'dtype': [str(t) for t in df.dtypes], 'missing': df.isna().sum().values})
```

```python
target = 'cd' if 'cd' in df.columns else next((c for c in df.columns if c.endswith('_cd')), None)
target
```

```python
if target:
    ax = df[target].dropna().hist(bins=30)
    ax.set_title(f'{target} distribution')
    ax.set_xlabel(target)
    ax.set_ylabel('count')
    plt.tight_layout()
    plt.savefig(FIG_DIR / 'target_distribution.png', dpi=160)
    plt.show()
```

```python
numeric = df.select_dtypes('number')
numeric.describe().T.sort_index()
```
